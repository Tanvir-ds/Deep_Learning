```python
# importing librraies
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating Log Loss function
def log_loss(y_true, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(x, epsilon) for x in y_predicted]
  y_predicted_new = [min(x, 1 - epsilon) for x in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  # y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
  # epsilon = 1e-15 = 0.000000000000001 <==> 1-epsilon = 0.999999999999999
  return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
# Creating sigmoid function
def sigmoid_numpy(X):
  import numpy
  return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([40,20,0,-43]))
```

```python
# custom Neural Network class
class myNN:
  # getting input parameters
  def __init__(self):
    self.w1 = 1
    self.w2 = 1
    self.bias = 0

  # custom function for fitting the model
  def fit(self, X, y, epochs, loss_threshold):
    self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X["affordability"], y, epochs, loss_threshold)

  # custom function for predictiction
  def predict(self, x_test):
    weighted_sum = self.w1 * x_test['age'] + self.w2 * x_test["affordability"] + self.bias
    return sigmoid_numpy(weighted_sum)

  # gradient descent function
  def gradient_descent(self, age, affordability , y_true, epochs, loss_threshold):

    # initializing w1, w2, bias
    w1 = w2 = 1
    bias = 0
    n = len(age)
    rate = 0.5

    for epoch in range(epochs):
      weighted_sum = w1*age + w2*affordability + bias
      y_predicted = sigmoid_numpy(weighted_sum)

      # defining loss
      loss = log_loss(y_true, y_predicted)

      # Updating w1, w2, bias based on d(deriavative)
      w1_d = (1/n)*np.dot(np.transpose(age), (y_predicted - y_true))
      w2_d = (1/n)*np.dot(np.transpose(affordability), (y_predicted - y_true))
      bias_d = np.mean(y_predicted - y_true)

      w1 = w1 - rate * w1_d
      w2 = w2 - rate * w2_d
      bias = bias - rate * bias_d

      if epoch % 50 == 0:
        print(f"Epoch:{epoch} w1:{w1} w2:{w2}, bias:{bias} loss:{loss}")

      if loss<=loss_threshold:
        print(f"Epoch:{epoch} w1:{w1} w2:{w2}, bias:{bias} loss:{loss}")
        break

    return w1, w2, bias
```

```python
# Calling the class
customModel = myNN()
customModel.fit(x_train, y_train, 500, 0.5509)
```
