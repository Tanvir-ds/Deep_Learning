```python
# importing libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np# Load the MNIST handwritten digits dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load the MNIST handwritten digits dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Standard Scaling
x_train = x_train/255
x_test = x_test/255
```

```python
# We have to flattening the dataset as input shape(28*28=784)
x_train_flatten = x_train.reshape(len(x_train), 28*28)
x_test_flatten = x_test.reshape(len(x_test), 28*28)
```

```python
# Shape of x_train_flatten
x_train_flatten.shape
```

```python
# Shape of x_train_flatten
x_test_flatten.shape
```

```python
x_train_flatten[0]
```

```python
x_train_flatten[0].shape
```

```python
# Training the model with single neural network
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation="sigmoid")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
    )

# Fitting the model
model.fit(x_train_flatten, y_train, epochs=5)
```

```python
# model Evaluation
model.evaluate(x_test_flatten, y_test)
```

```python
# Prediction of test dataset
y_pred = model.predict(x_test_flatten)
```

```python
# Visualizing first x_test dataset
plt.matshow(x_test[0])
```

```python
# Actual result of first x_test dataset
y_test[0]
```

```python
# Predicted result of first x_test dataset
np.argmax(y_pred[0])
```

```python
# Labels of y_pred
y_pred_labels = [np.argmax(x) for x in y_pred]
y_pred_labels[:5]
```

```python
# Confusion matrix
cm = tf.math.confusion_matrix(y_test, y_pred_labels)
cm
```

```python
# Visulaizing confusion matrix
plt.figure(figsize=(10,7))
sns.set()
sns.heatmap(cm, annot=True, cbar=True, fmt=".1f", cmap="viridis")
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```

```python
# Adding multiple neural network
# Training the model with multiple neural network
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
    )

# Fitting the model
model.fit(x_train_flatten, y_train, epochs=5)
```

```python
# model Evaluation
model.evaluate(x_test_flatten, y_test)
```

```python
# Creating confusion matrix
y_pred = model.predict(x_test_flatten)
y_pred_labels = [np.argmax(x) for x in y_pred]
cm = tf.math.confusion_matrix(y_test, y_pred_labels)
plt.figure(figsize=(10,7))
sns.set()
sns.heatmap(cm, annot=True, cbar=True, fmt=".1f")
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
plt.show()
```

```python
# Add Flattening in model training this will give same result like before
# Training the model with multiple neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
    )

# Fitting the model
model.fit(x_train, y_train, epochs=5)
```

```python
# model Evaluation
model.evaluate(x_test, y_test)
```

```python
# Creating confusion matrix
y_pred = model.predict(x_test)
y_pred_labels = [np.argmax(x) for x in y_pred]
cm = tf.math.confusion_matrix(y_test, y_pred_labels)
plt.figure(figsize=(10,7))
sns.set()
sns.heatmap(cm, annot=True, cbar=True, fmt=".1f", cmap="Blues")
plt.xlabel("Predicted Value")
plt.ylabel("Actual Value")
```
