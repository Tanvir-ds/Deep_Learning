```python
# importing librraies
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset
df = pd.read_excel("/content/simple_insurance.xlsx")
df
```

```python
# Extracting Dependent and Independent variable
x = df.drop(["insurance"], axis=1)
y = df["insurance"]
```

```python
x
```

```python
# Scaling age column value from 0 to 1
x["age"] = x["age"]/100
x
```

```python
y
```

```python
# Splitting the dataset into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2023)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
x_test
```

```python
# Training the model on tensorflow
# Initializing co-efficient(w1,w2) as 1 and intercept as 0
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation="sigmoid", kernel_initializer="ones", bias_initializer="zeros")
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# Fitting the model
model.fit(x_train, y_train, epochs=5000)
```

```python
# Evaluate the model
model.evaluate(x_test, y_test)
```

```python
# Prediction of test dataset
y_predicted = model.predict(x_test)
y_predicted
```

Gradient Descent from Scratch

```python
# creating Log Loss function
def log_loss(y_true, y_predicted):
  epsilon = 1e-15
  y_predicted_new = [max(x, epsilon) for x in y_predicted]
  y_predicted_new = [min(x, 1 - epsilon) for x in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  # y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
  # epsilon = 1e-15 = 0.000000000000001 <==> 1-epsilon = 0.999999999999999
  return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
# Creating sigmoid function
def sigmoid_numpy(X):
  import numpy
  return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([40,20,0,-43]))
```

```python
# gradient descent
def gradient_descent(age, affordability , y_true, epochs, loss_threshold):

  # initializing w1, w2, bias
  w1 = w2 = 1
  bias = 0
  n = len(age)
  rate = 0.5

  for i in range(epochs):
    weighted_sum = w1*age + w2*affordability + bias
    y_predicted = sigmoid_numpy(weighted_sum)

    # defining loss
    loss = log_loss(y_true, y_predicted)

    # Updating w1, w2, bias based on d(deriavative)
    w1_d = (1/n)*np.dot(np.transpose(age), (y_predicted - y_true))
    w2_d = (1/n)*np.dot(np.transpose(affordability), (y_predicted - y_true))
    bias_d = np.mean(y_predicted - y_true)

    w1 = w1 - rate * w1_d
    w2 = w2 - rate * w2_d
    bias = bias - rate * bias_d

    print(f"Epoch:{i} w1:{w1} w2:{w2}, bias:{bias} loss:{loss}")

    if loss<=loss_threshold:
      break

  return w1, w2, bias
```

```python
import numpy as np

# Log Loss Function
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))

# Sigmoid Function
def sigmoid_numpy(X):
    return 1 / (1 + np.exp(-X))

# Gradient Descent Function
def gradient_descent(X, y_true, epochs, loss_threshold):
    n_samples, n_features = X.shape
    rate = 0.5

    # Initializing weights and bias
    w = np.ones(n_features)
    bias = 0

    for epoch in range(epochs):
        weighted_sum = np.dot(X, w) + bias
        y_predicted = sigmoid_numpy(weighted_sum)

        # Calculating loss
        loss = log_loss(y_true, y_predicted)

        # Calculating gradients
        w_gradients = (1/n_samples) * np.dot(X.T, (y_predicted - y_true))
        bias_gradient = np.mean(y_predicted - y_true)

        # Updating weights and bias
        w -= rate * w_gradients
        bias -= rate * bias_gradient

        print(f"Epoch:{epoch} Loss:{loss}")

        if loss <= loss_threshold:
            break

    return w, bias

# Assuming you have x_train and y_train as input data
# Assuming x_train has shape (n_samples, n_features)
x_train = np.array(x_train)  # Your training data
y_train = np.array(y_train)  # Your binary labels

# Performing gradient descent
optimal_w, optimal_bias = gradient_descent(x_train, y_train, epochs=1000, loss_threshold=0.5469)
print(f"w1:{optimal_w[0]}, w2:{optimal_w[1]}, bias:{optimal_bias}")
```
